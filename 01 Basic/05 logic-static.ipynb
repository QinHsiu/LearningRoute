```python
import numpy as np
import pandas as pd
import os
data=pd.read_csv(r'Customer-Churn.csv')
data.columns
data.dtypes
```

```python
var_name=['MonthlyCharges','tenure']
np.min(data[var_name],axis=0)
```

```python
#获取相关数据
data.describe(include=['object'])
data.describe()
```

```python
#分组
groupby=data.groupby('MonthlyCharges')
groupby.max()
groupby.size
groupby.cumcount()
```

```python
name=['MonthlyCharges','tenure']
group=data[name].groupby('MonthlyCharges')
data[name]
group.mean()
```

```python
#分组聚合agg
import numpy as np
group.agg([np.mean,np.sum])
```

```python
group.agg([np.mean,np.sum]).loc[[18.25,118.20],('tenure',['mean','sum'])]
```

```python
#自定义函数
def double(data):
    return data.sum()*2
group.agg({'tenure':double})
```

```python
data
pd.pivot_table(data,index='tenure',values='MonthlyCharges',aggfunc=np.mean,margins=True,margins_name='月费用求均值')
pd.pivot_table(data,index='tenure',columns='TotalCharges',values='MonthlyCharges',fill_value=0,aggfunc=np.mean,margins=True,margins_name='月费用求均值')
```

```python
#交叉表,计算频数,百分比
pd.crosstab(index=data['MonthlyCharges'],columns=['tenure'],margins=True,normalize='columns')
```
